## Exercise 1.3

Write a script that will convert an arbitrary integer to a binary number (the number will be represented as a string with only *0* and *1*). Avoid using the *bin()* function or any other pythons default functions

```python
def binary(num):
    if num == 0:
        return "0"

    negative = num < 0
    n = abs(num)
    result = ""

    while n > 0:
        result = str(n % 2) + result
        n = n // 2

    return "-" + result if negative else result

#For the negative number may be more correct to do two's complement.

print(binary(0))
print(binary(5))
print(binary(-13))

#To check:
print(bin(0))
print(bin(5))
print(bin(-13))
```

## Exercise 1.4 - The Fibonacci Sequence:

The Fibonacci Sequence is a series of numbers. The following number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13. The following number in this series above is 8 + 13 = 21


Your task is to implement a function *fibonacci* that takes an integer as an input and returns a list that contains all Fibonacci numbers with values lower than the input integer

Example:

*print(fibonaci(10))*

[0, 1, 1, 2, 3, 5, 8]


```python
def fibonacci(upper_threshold: int) -> list:
    if upper_threshold <= 0:
        return []

    fib_list = [0]
    if upper_threshold > 1:
        fib_list.append(1)

    while True:
        next_fib = fib_list[-1] + fib_list[-2]
        if next_fib >= upper_threshold:
            break
        fib_list.append(next_fib)

    return fib_list

print(fibonacci(10))
```

#### Advanced

Try to implement the function using multiple approaches: *Iterative approach*, *Recusrion*, *Memoization*

```python
def fibonacci_iterative(upper_threshold: int) -> list:
    #Iterative approach
    if upper_threshold <= 0:
        return []
    a, b = 0, 1
    result = [a]
    if upper_threshold > 1:
        result.append(b)
    while True:
        a, b = b, a + b
        if b >= upper_threshold:
            break
        result.append(b)

    return result

def fibonacci_recursive(upper_threshold: int) -> list:
    #Recursive approach
    def fib(n):
        if n <= 1:
            return n
        return fib(n-1) + fib(n-2)
    result2 = []
    i = 0
    while True:
        val = fib(i)
        if val >= upper_threshold:
            break
        result2.append(val)
        i += 1
    return result2

def fibonacci_memoization(upper_threshold: int) -> list:
    #Memoization approach with cache
    cache = {}
    def fib_memo(n):
        if n in cache:
            return cache[n]
        if n <= 1:
            cache[n] = n
        else:
            cache[n] = fib_memo(n-1) + fib_memo(n-2)
        return cache[n]
    result3 = []
    i = 0
    while True:
        val = fib_memo(i)
        if val >= upper_threshold:
            break
        result3.append(val)
        i += 1
    return result3

print(fibonacci_iterative(10))
print(fibonacci_recursive(10))
print(fibonacci_memoization(10))
```

## Exercise 1.5. -  Rock, Paper, Scissors game:

#### basic

*Rock, Paper, Scissors* is a well-known and straightforward game. If you do not know the rules, google them. We will write a code for the Rock, Paper, Scissors game where the user plays against a random computer. The code can be written into one function *rock_paper_scissors*.

Notes:
- This implementation uses the **random** library to enable the computer to make a random choice.
- Keyword **Input** may be helpful. Check it
- After the function is run, the program will ask you about your movement, which you type in
- The game is case-insensitive for user input (e.g., "rock", "Rock", and "ROCK" are all valid).
- This script plays one round of the game.
- The script prints the result ('You lose', 'You win', 'It is a tie')


```python
import random

def rock_paper_scissors() -> None:
    # Computer makes a random choice
    choices = ["rock", "paper", "scissors"]
    computer = random.choice(choices)

    # User decides
    user_input = input("Choose rock, paper or scissors: ").strip()

    # Case-insensitive
    user = user_input.lower()

    # Check if input is valid
    if user not in choices:
        print("Invalid choice")
        return

    # Show what both players chose
    print(f"You chose: {user_input}")
    print(f"Computer chose: {computer}")

    # Determine the result
    if user == computer:
        print("It is a tie")
    elif (user == "rock" and computer == "scissors") or \
         (user == "paper" and computer == "rock") or \
         (user == "scissors" and computer == "paper"):
        print("You win")
    else:
        print("You lose")

# Run the game
rock_paper_scissors()
```

#### Advanced

Extend the Rock, Paper, Scissors game to be able to play *n* rounds.

Notes:
- The scrips will contain two counters - user score and computer score
- After each run, the script prints the current scores and rounds.
- When the number of rounds is reached, the script prints the result ('You lose', 'You win', 'It is a tie')

```python
import random

def rock_paper_scissors_advanced(rounds: int) -> None:

    choices = ["rock", "paper", "scissors"]
    user_score = 0
    computer_score = 0

    print(f"=== Best of {rounds} rounds ===\n")

    for round_num in range(1, rounds + 1):
        print(f"--- Round {round_num} of {rounds} ---")

        # Computer makes a random choice
        computer = random.choice(choices)

        # Ask user
        user_input = input("Choose rock, paper or scissors: ").strip()
        user = user_input.lower()

        # Check if input is valid
        if user not in choices:
            print("Invalid choice")
            continue

        # Show what both players chose
        print(f"You chose: {user_input}")
        print(f"Computer chose: {computer}")

        # Determine the result
        if user == computer:
            print("It is a tie")
        elif (user == "rock" and computer == "scissors") or \
             (user == "paper" and computer == "rock") or \
             (user == "scissors" and computer == "paper"):
            print("You win this round!")
            user_score += 1
        else:
            print("Computer wins this round")
            computer_score += 1

        # Show current score
        print(f"Score: You {user_score} - {computer_score} Computer\n")

    # Final result
    print("=== GAME OVER ===")
    print(f"Final Score: You {user_score} - {computer_score} Computer")

    if user_score > computer_score:
        print("YOU WIN THE GAME! ")
    elif user_score < computer_score:
        print("COMPUTER WINS THE GAME")
    else:
        print("IT'S A TIE GAME")

rock_paper_scissors_advanced(3)
```

## Exercise 2.1


In this exercise, you will work with NumPy arrays and learn how to efficiently modify and process them using both loops and vectorized operations. Your task is to:

1. Implement the function *create_array_nxn* that generates an n×n NumPy array filled with numbers from n² - 1 down to 0
2. Using a loop-based approach (for-loop, while-loop), implement the function *apply_threshold_loop* that takes any NumPy array and replaces all numbers below a user-defined threshold with 0.
3. Implement function *apply_threshold_vectorized* that performs the same operation as *apply_threshold_loop*, but without loops, using NumPy vectorized operations
4. Implement function *compare_performance* that compares the execution time of functions *apply_threshold_vectorized* and *apply_threshold_loop* with **time** library and print the results


**Additional Notes**:

- Ensure n is a positive integer (n > 0).
- Use the time library to measure execution times for performance comparison.
- The function *compare_performance(n, threshold)* should:
    - Generate an n×n array.
    - Apply both loop-based and vectorized thresholding.
    - Print execution times.

```python
import numpy as np
import time
```

```python
def create_array_nxn(n: int) -> np.ndarray:
    # Create array with numbers from n²-1 down to 0
    arr = []
    num = n*n - 1
    for i in range(n):
        row = []
        for j in range(n):
            row.append(num)
            num -= 1
        arr.append(row)
    return np.array(arr)
```

```python
def apply_threshold_loop(arr: np.ndarray, threshold: int) -> np.ndarray:
    result = arr.copy()
    for i in range(len(result)):
        for j in range(len(result[i])):
            if result[i][j] < threshold:
                result[i][j] = 0
    return result
```

```python
def apply_threshold_vectorized(arr: np.ndarray, threshold: int) -> np.ndarray:
    result = arr.copy()
    result[result < threshold] = 0
    return result
```

```python
def compare_performance(n: int, threshold: int) -> None:
    arr = create_array_nxn(n)

    start = time.time()
    loop_result = apply_threshold_loop(arr, threshold)
    loop_time = time.time() - start

    start = time.time()
    vec_result = apply_threshold_vectorized(arr, threshold)
    vec_time = time.time() - start

    print(f"n = {n}")
    print(f"Loop method: {loop_time:.6f} seconds")
    print(f"Vectorized method: {vec_time:.6f} seconds")
    print(f"Vectorized is {loop_time/vec_time:.2f}x faster")

# Test
compare_performance(500, 10000)
```

```python
# Exercise 3 - California Housing Dataset
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
dataset = pd.read_csv('sample_data/california_housing_test.csv')

# 1. Check what describe() does
print("\n1. dataset.describe():")
print(dataset.describe())

# 2. Display rows where total_bedrooms > 310
print("\n2. Rows with total_bedrooms > 310:")
print(dataset[dataset['total_bedrooms'] > 310])

# 3. Drop first and last row
dataset = dataset.iloc[1:-1]
print("\n3. First and last row dropped")

# 4. Plot households with mean
mean_hh = dataset['households'].mean()
plt.figure(figsize=(10, 5))
plt.scatter(range(len(dataset)), dataset['households'], c='blue', alpha=0.5)
plt.axhline(y=mean_hh, color='red', linewidth=2)
plt.xlabel('Index')
plt.ylabel('Households')
plt.title('Households with Mean Line')
plt.show()

# 5. Check for NaN values and replace
print("\n5. NaN values before:", dataset.isna().sum().sum())
dataset = dataset.fillna(dataset.mean())
print("   NaN values after:", dataset.isna().sum().sum())

# 6. Plot latitude, longitude
plt.figure(figsize=(8, 6))
plt.scatter(dataset['latitude'], dataset['longitude'], c='green', alpha=0.5, s=5)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Latitude and Longitude')
plt.show()

# 7. Min-max normalization
col1, col2 = 'total_rooms', 'population'
dataset[col1 + '_norm'] = (dataset[col1] - dataset[col1].min()) / (dataset[col1].max() - dataset[col1].min())
dataset[col2 + '_norm'] = (dataset[col2] - dataset[col2].min()) / (dataset[col2].max() - dataset[col2].min())
print(f"\n7. Normalized {col1} and {col2} (first 5 rows):")
print(dataset[[col1 + '_norm', col2 + '_norm']].head())

# 8. Correlation Matrix
print("\n8. Correlation Matrix:")
corr = dataset.select_dtypes(include=[np.number]).corr()
print(corr)

plt.figure(figsize=(8, 6))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()

```
